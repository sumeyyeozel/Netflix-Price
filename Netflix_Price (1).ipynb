# Statistiken

```python
features=['Total Library Size', 'Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)']
ax = sns.heatmap(data = df[features].corr(),cmap='YlGnBu',annot=True)
```

Wenn die hheren Werte der einen Variablen in der Tendenz mit den niedrigeren Werten der anderen korrelieren, besteht eine negative oder inverse Beziehung zwischen den beiden Variablen.
In diesem Korrelation zum Beispiel zwischen Cost Per Month-Basic und Total Library Size haben negative Korrelation.
Ich dachte, dass es zwischen Total Library Size and Cost das hohe Korrelation gibt. Das Heatmap hat uns jedoch gezeigt, dass es keine hohe Korrelation zwischen diesen beiden Merkmalen gibt

```python
fig, ax=plt.subplots(figsize=(10,5))
plt.scatter(df['Cost Per Month - Basic ($)'], df['Cost Per Month - Standard ($)'])
ax.set_xlabel('Basic')
ax.set_ylabel('Standard')
plt.show()
```

Whrend die Basic Kosten steigen, steigen die Standard Kosten. Weil die Beziehung positiv ist, lautet man diese Beziehung die positive Korrelation.

```python
fig, ax=plt.subplots(figsize=(10,5))
plt.scatter(df['Cost Per Month - Premium ($)'], df['Total Library Size'])
ax.set_xlabel('Cost')
ax.set_ylabel('Library Size')
plt.show()
```

Whrend die Kosten steigen, senkt die Gre der Bibliothek. Weil die Beziehung negativ ist, lautet man diese Beziehung die negative Korrelation. 

# Modelling

**Linear** **Regression**

**Y = a + bX ** 
a=intercept
b=slope


```python
from scipy import stats
x=df['Cost Per Month - Basic ($)']
y=df['Cost Per Month - Standard ($)']
slope, intercept, r, p, std_err = stats.linregress(x,y)
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
